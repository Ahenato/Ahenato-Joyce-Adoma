```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
red_wine=pd.read_csv(filepath_or_buffer="red_wine.csv")
head=red_wine.head(10)
tail=red_wine.tail(20)
```

### Alcohol Content

```python
HeadGraph_Alcohol=head["alcohol"]
plt.xlabel("Alcohol Content")
plt.ylabel("Freqency")
plt.hist(HeadGraph_Alcohol)
```

### Alcohol Content Against Wine Quality

```python

HeadGraph_Quality=head["quality"]
plt.scatter(HeadGraph_Alcohol,HeadGraph_Quality,alpha=0.5)
plt.title('Alcohol Content vs. Wine Quality')
plt.xlabel('Alcohol Content')
plt.ylabel('Quality')
plt.grid(True)
```

### Quality Against Fixed Acidity

```python
plt.bar(head['quality'],head['fixed acidity'])
plt.xlabel('Quality')
plt.ylabel('Fixed acidity')
plt.title("Quality Against Fixed Acidity")
```

### Quality Against Alcohol

```python
plt.bar(head['quality'],head['alcohol'])
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.title("Quality Against Alcohol")
```

### Heat Map

```python
sb.heatmap(head.corr(), annot =True, fmt = '0.1f')
```
